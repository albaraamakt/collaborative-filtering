```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
df = pd.read_csv('food_data_1.csv')
df
```

```python
# Create a TfidfVectorizer
tfidf = TfidfVectorizer()

# Fit and transform the data to a tfidf matrix
tfidf_matrix = tfidf.fit_transform(df['ingredients'])

# Print the shape of the tfidf_matrix
tfidf_matrix.shape
```

```python
# Compute the cosine similarity between each movie description
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(df.index, index=df['name']).drop_duplicates()
indices
```

```python
def get_recommendations(name, cosine_sim=cosine_sim, num_recommend=10):
    idx = indices[name]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    top_similar = sim_scores[1:num_recommend+1]
    meal_indices = [i[0] for i in top_similar]
    
    return df['name'].iloc[meal_indices]
```

```python
get_recommendations('Tavuk Döner', num_recommend=10)
```
