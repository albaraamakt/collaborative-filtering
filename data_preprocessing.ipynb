```python
import random
```

```python
min(df1['price'])
```

```python
orders1 = [{'order_id': i,
            'user_id': 0, 
            'meal_id': random.choice(df1['id']),
            'quantity': random.randint(1, 5),
        } for i in range(100)]
```

```python
orders1[1:10]
```

```python
orders2 = [{'order_id': i + 100,
            'user_id': 1, 
            'meal_id': random.choice(df1['id']),
            'quantity': random.randint(1, 5),
        } for i in range(100)]
```

```python
orders3 = [{'order_id': i + 100 * 2,
            'user_id': 2, 
            'meal_id': random.choice(df1['id']),
            'quantity': random.randint(1, 5),
        } for i in range(100)]
```

```python
orders4 = [{'order_id': i + 100 * 3,
            'user_id': 3, 
            'meal_id': random.choice(df1['id']),
            'quantity': random.randint(1, 5),
        } for i in range(100)]
```

```python
orders = []
orders.extend(orders1)
orders.extend(orders2)
orders.extend(orders3)
orders.extend(orders4)
```

```python
orders_table = pd.DataFrame(orders)
```

```python
orders_table['order_id'].duplicated
```

```python
orders_table.loc[100]
```

```python
del orders_table['quantity']
```

```python
orders_table
```

```python
user0_orders = orders_table[orders_table['user_id'] == 0]
user0_orders
```

```python
user0_orders = pd.merge(df1, user0_orders, on='meal_id', how='inner')
```

```python
user0_orders.to_csv('user0_orders', index=False)
```

```python

for i in range(0, 4):
    user_orders = orders_table[orders_table['user_id'] == i]
    user_orders = pd.merge(df1, user_orders, on='meal_id', how='inner')
    user_orders.to_csv(f'user{i}_orders.csv', index=False)
```
